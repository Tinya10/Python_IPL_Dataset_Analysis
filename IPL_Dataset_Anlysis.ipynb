```python
import yaml
```

```python
with open('G:/Data Science ML AI/Data course files/ipl_match.txt') as f:
    data = yaml.load(f)
```

```python
data
```

```python
type_of_data = type(data)
print("The data type that you are working with is: ",type_of_data)
```

```python
city = data['info']['city']
print("The match was played in",city)
```

```python
venue = data['info']['venue']
print("The match was played at:",venue)
```

```python
Teams = data['info']['teams']
print("The teams that participated in the tournament were:" + Teams[0]  +' and ' + Teams[1])
print("There are total " + str(len(Teams)) + " teams played in the tournament")
```

```python
toss = data['info']['toss']['winner'], data['info']['toss']['decision']
print("The team that won the toss is " + toss[0] + " and they decided to " + toss[1])
```

```python
first_bowler = data['innings'][0]['1st innings']['deliveries'][0][0.1]['bowler']
first_batsman = data['innings'][0]['1st innings']['deliveries'][0][0.1]['batsman']
print("The bowler who bowled the first ball is ",first_bowler)
print("The batsman who bat the first ball is ",first_batsman)
```

```python
deliveries = len(data['innings'][0]['1st innings']['deliveries'])
print("Total deliveries in first innings were ",deliveries)
```

```python
deliveries_2 = len(data['innings'][1]['2nd innings']['deliveries'])
print("Total deliveries in second innings were ",deliveries_2)
```

```python
winning_team = data['info']['outcome']['winner']
by_margin = data['info']['outcome']['by']['runs']
print("The winning team was " + winning_team + " by " + str(by_margin) + " runs")
```
